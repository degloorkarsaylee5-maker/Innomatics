```python
class smart_light:
    
    def __init__(self,light_name):
        self.name=light_name
        self.status="off"
    def turn_on(self):
        self.status="On"
    def turn_off(self):
        self.status="off"
    def display_status(self):
        print(f"{self.name} is {self.status}")

light_name=input("Enter light name:")
action=input("On or off? :- ")

light=smart_light(light_name)

if action=='On':
    light.turn_on()
elif action=='off':
    light.turn_off()
light.display_status()
    
```

```python
class IDcard:
    
    def __init__(self, employee_name, emp_id, dept):
        self.name = employee_name
        self.id = emp_id
        self.dept = dept
    
    def display_details(self):
        print("Employee ID card:")
        print(f"Name:- {self.name}")
        print(f"ID:- {self.id}")
        print(f"Department:- {self.dept}")


employee_name = input("Enter name: ")
emp_id = input("Id:- ")
dept = input("Enter department:- ")

detail = IDcard(employee_name, emp_id, dept)
detail.display_details()
```

```python
class Contact:
    
    def __init__(self, name, mob_no):
        self.name = name
        self.mob_no = mob_no
    
    def display_details(self):
        print("Contact Saved")
        print(f"Name:- {self.name}")
        print(f"Phone number:- {self.mob_no}")


name = input("Enter name: ")
mob_no = int(input("Number:- "))


detail = Contact(name, mob_no)
detail.display_details()
```

```python
class Price:
    
    def __init__(self, product_name,prod_price):
        self.name = product_name
        self.price = prod_price
    
    def display_details(self):
        
        print(f"Product Name:- {self.name}")
        print(f"Product Price:- ₹{self.price}")



name = input("Enter name: ")
price = int(input("Number:- "))


detail = Price(name, price)
detail.display_details()
```

```python
class Movie:
    
    def __init__(self, movie_name,movie_rating):
        self.name = movie_name
        self.price = movie_rating

    def display_details(self):
        
        print(f"Movie Name:- {self.name}")
        print(f"Movie Rating:- {self.price}/5")


name = input("Enter name: ")
rating = float(input("Rating:- "))


detail = Movie(name, rating)
detail.display_details()
```

```python
class Delivery:
    
    def __init__(self, customer_name,address):
        self.name = customer_name
        self.address = address

    def display_details(self):
        print("Delivery Details")
        print(f"Customer:- {self.name}")
        print(f"Address:- {self.address}")


name = input("Enter your name: ")
address = input("Address:- ")


detail = Delivery(name, address)
detail.display_details()
```
