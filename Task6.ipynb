```python
def parking(capacity, logs):
    curr = 0
    peak = 0
    
    for log in logs:
        if log == "IN":
            curr+= 1
        elif log == "OUT":
            if curr > 0:
                curr -= 1
        
        # Track peak usage
        if curr > peak:
            peak = curr
    
    print("Currently Parked Vehicles:", curr)
    print("Peak Parking Usage:", peak)
    
    if curr > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")

capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking(capacity, logs)
```

```python
def delivery_time(distance, traffic, weather):
    speed = 40 
    
    time = (distance / speed) * 60
    
    delay = 0
    
    if traffic == "Low":
        delay += 0
    elif traffic == "Medium":
        delay += 10
    elif traffic == "High":
        delay += 20
    
    if weather == "Clear":
        delay += 0
    elif weather == "Rainy":
        delay += 15
    elif weather == "Storm":
        delay += 25
    
    eta = time + delay
    
    print("Estimated Delivery Time:", round(eta), "minutes")

delivery_time(8, "High", "Rainy")
```

```python
def occupancy(seats, booked):
    booked_count = len(booked)
    
    occupancy = (booked_count /seats) * 100
    
    print("Occupancy:", round(occupancy), "%")
    
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open Additional Show")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

total_seats = 200
booked_seats = [1] * 150  

occupancy(total_seats, booked_seats)
```

```python
def server_load(readings):
    total = 0
    
    for reading in readings:
        total += reading
    
    average = total / len(readings)
    
    print("Average CPU Load:", round(average), "%")
    
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
    
cpu = [45, 60, 70, 85, 90]
server_load(cpu)
```

```python
def classroom_usage(resource_usage):
    
    # predefined safe usage limits
    limits = {
        "Projector": 8,
        "AC": 8,
        "Lights": 10
    }
    
    overused = []
    
    for resource in resource_usage:
        if resource_usage[resource] > limits.get(resource, 0):
            overused.append(resource)
    
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

usage = {
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

classroom_usage(usage)
```

```python
def event_registration(capacity, registrations):
    
    confirmed = 0
    waitlist = 0
    
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    
    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

event_registration(100, 105)
```
