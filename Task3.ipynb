```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

max_value = max(employees.values())

top_employees = []

for name, value in employees.items():
    if value == max_value:
        top_employees.append(name)

print(f"Top Performers Eligible for Bonus: {top_employees} (Score: {max_value})")
```

```python
query = input("Enter the query: ").lower()

words = query.split()
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# store only repeated words 
repeated_words = {}
for word, count in word_count.items():
    if count > 1:
        repeated_words[word] = count

print(repeated_words)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):#enumerate used so that we could get the indexed list
    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

count = {}
for email in emails:
    domain = email.split("@")[1]
    if domain in count:
        count[domain] += 1
    else:
        count[domain] = 1
        
user_count = len(emails)

for domain, count in count.items():
    percentage = (count / user_count) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg = sum(sales) / len(sales)

more_than_avg = avg * 1.30

for index, value in enumerate(sales):
    if value > more_than_avg:
        print(f"Day {index + 1}: {value}")
```
