```python
def validate_recharge(amount):
    vaid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least â‚¹50.")
        return False

    if amount in vaid_plans:
        return True
    else:
        print("Invalid Plan! Valid plans are 199, 299, 399, 599.")
        return False


while True:
    try:
        ammount = int(input("Enter recharge amount: "))
        
        if validate_recharge(ammount):
            print("Recharge Successful!")
            break   
        else:
            print("Please try again.\n")

    except ValueError:
        print("Invalid input! Please enter a numeric value.\n")
```

```python
def check_stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")

produccts = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 25,
    "Salt": 12
}

check_stock(produccts)
```

```python
def calculate_fare(distance, is_peak):
    base_farre = 50
    per_km_rate = 12

    fae = base_farre + (distance * per_km_rate)

    if is_peak:
        fae += fae * 0.25 #adding 25 % extra

    return fae

while True:
    try:
        distancce = float(input("Enter distance in km: "))
        peak_nput = input("Is it peak hour? (yes/no): ").lower()

        if peak_nput == "yes":
            iss_peak = True
        elif peak_nput == "no":
            iss_peak = False
        else:
            print("Invalid input for peak hour. Please enter yes or no.\n")
            continue

        tootal_fare = calculate_fare(distancce, iss_peak)

        print("Total Fare: ", tootal_fare)

        rtry = input("Do you want to calculate again? (yes/no): ").lower()
        if rtry != "yes":
            print("Fare calculated.")
            break

    except ValueError:
        print("Invalid input!\n")
```

```python
def check_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_count = 0

    for status in attendance_list:
        if status == "P":
            present_count += 1

    attendance_percentage = (present_count / total_days) * 100

    if attendance_percentage >= 75:
        return attendance_percentage, "Eligible"
    else:
        return attendance_percentage, "Not Eligible"
    
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

percentage, result = check_eligibility(attendance)

print("Attendance Percentage:", percentage)
print("Status:", result)
```

```python
def check_password(password):
    has_digit = False
    ha_special = False
    specal_chars = "@#$"

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in specal_chars:
            ha_special = True

    if has_digit and ha_special:
        return "Strong Password"
    else:
        return "Weak Password(add special characters)"

pssorwd = input("Enter password: ")
ressult = check_password(pssorwd)
print(ressult)
```
