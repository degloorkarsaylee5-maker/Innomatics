Q1. Difference between "Love" and "love" in NLP?
Computers compare text character by character, so "Love" and "love" are treated as two completely different tokens. The capital "L" makes them distinct. This is why lowercasing is one of the first preprocessing steps — it maps both words to a single token love.

Q2. What happens if stopwords are not removed?
Stopwords like "is", "the", "a", "am" appear very frequently but carry almost no meaning. Without removing them they dominate word frequency counts and skew analysis machine learning models pick up irrelevant patterns computational cost increases unnecessarily

Q3. Two scenarios where removing stopwords is harmful
Case 1 – Sentiment Analysis:  
This is not good. → removing "not" flips the meaning to positive. Very dangerous
Case 2 – Question Answering:**
"What is the capital of India?" → removing "what", "is", "the" destroys the question structure.

Q4. Stemming vs Lemmatization

Stemming and lemmatization are both techniques used to reduce words to their base form, but they work quite differently.

Stemming is a simpler and faster method. It works by chopping off word endings using basic rules, without considering the actual meaning of the word. Because of this, the result may not always be a valid word. For example, the word "studies" might be reduced to "studi," which isn’t a real word.

Lemmatization, on the other hand, is more advanced and accurate. It uses vocabulary and grammatical analysis to convert a word into its proper base or dictionary form. This means the output is always a meaningful word. For instance, "studies" would be converted to "study."

In terms of performance, stemming is faster but less accurate, while lemmatization is slower but provides more precise and meaningful results.



## Task 2: Build Advanced Preprocessing Function


```python
import re

MEANINGFUL_SHORT_WORDS = {"no", "not"}

def preprocess_text(raw_text):
    if not isinstance(raw_text, str) or raw_text.strip() == "":
        return []

    cleaned = raw_text
    cleaned = re.sub(r"http\S+|https\S+|www\.\S+|\S+@\S+\.\S+", "", cleaned)
    cleaned = cleaned.encode("ascii", "ignore").decode("ascii")
    cleaned = cleaned.lower()
    cleaned = re.sub(r"\d+", "", cleaned)
    cleaned = re.sub(r"[^a-z\s]", "", cleaned)
    cleaned = re.sub(r"(.)\1{2,}", r"\1", cleaned)
    cleaned = re.sub(r"\s+", " ", cleaned).strip()
    word_tokens = cleaned.split()
    filtered_tokens = [
        word for word in word_tokens
        if len(word) > 2 or word in MEANINGFUL_SHORT_WORDS
    ]

    return filtered_tokens

sample = "WOW!!! I looooved this product Visit http://example.com now!!!"
result = preprocess_text(sample)
print("Input  :", sample)
print("Tokens :", result)
print("Output :", " ".join(result))
```

## Task 3: Stress Testing


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print(" STRESS TEST RESULTS")

for idx, sentence in enumerate(test_sentences, start=1):
    cleaned_tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(cleaned_tokens)
    print(f"\n[{idx}] Original Text  : {sentence}")
    print(f"    Cleaned Tokens  : {cleaned_tokens}")
    print(f"    Cleaned Sentence: {cleaned_sentence}")

    

```

## Task 4: Token Analytics


```python
def compute_token_stats(sentence):
    token_list   = preprocess_text(sentence)
    total_count  = len(token_list)
    unique_count = len(set(token_list))
    avg_length   = round(sum(len(t) for t in token_list) / total_count, 2) if total_count > 0 else 0
    return {"sentence": sentence, "tokens": token_list,
            "total": total_count, "unique": unique_count, "avg_length": avg_length}


print("TOKEN ANALYTICS")


analytics_results = []
for sentence in test_sentences:
    stats = compute_token_stats(sentence)
    analytics_results.append(stats)
    print(f"\nSentence      : {stats['sentence']}")
    print(f"Total Tokens  : {stats['total']}")
    print(f"Unique Tokens : {stats['unique']}")
    print(f"Avg Token Len : {stats['avg_length']}")
```

```python
# Which sentence had the most noise?
noise_scores = []
for sentence in test_sentences:
    original_count = len(sentence.split())
    cleaned_count  = len(preprocess_text(sentence))
    noise_ratio    = round((original_count - cleaned_count) / original_count * 100, 1)
    noise_scores.append((sentence, noise_ratio))

noise_scores.sort(key=lambda x: x[1], reverse=True)

print("Sentences ranked by noise level:")
for rank, (sent, ratio) in enumerate(noise_scores, start=1):
    print(f"  {rank}. [{ratio}% noise] {sent}")

print("\nMost noisy sentence   :", noise_scores[0][0])
print("Most meaningful (least noise):", noise_scores[-1][0])
```


## Task 5: Frequency Analysis


```python
from collections import Counter

all_cleaned_tokens = []
for sentence in test_sentences:
    all_cleaned_tokens.extend(preprocess_text(sentence))

token_frequency = Counter(all_cleaned_tokens)

print("   TOP 10 MOST FREQUENT WORDS")
for word, freq in token_frequency.most_common(10):
    bar = "." * freq
    print(f"  {word:<15} {freq:>3}  {bar}")

print("   TOP 5 LEAST FREQUENT WORDS")
least_common = token_frequency.most_common()[:-6:-1]
for word, freq in least_common:
    bar = "*" * freq
    print(f"  {word:<15} {freq:>3}  {bar}")
```


## Task 6: Full Pipeline


```python
def full_pipeline(text_list):
    collected_tokens    = []
    processed_sentences = []
    for raw_text in text_list:
        token_output = preprocess_text(raw_text)
        collected_tokens.extend(token_output)
        processed_sentences.append(" ".join(token_output))
    return {"tokens": collected_tokens, "clean_sentences": processed_sentences}


pipeline_result = full_pipeline(test_sentences)

print("   FULL PIPELINE OUTPUT")
print("\nAll Tokens:")
print(pipeline_result["tokens"])
print("\nCleaned Sentences:")
for i, s in enumerate(pipeline_result["clean_sentences"], start=1):
    print(f"  [{i}] {s}")
```


## Task 7: Error Handling


```python
edge_case_inputs = [
    "",               # Empty string
    "1234 5678 999",  # Only numbers
    "!!! $$$ ???",    # Only special characters
    "   ",            # Only whitespace
]

print("EDGE CASE / ERROR HANDLING TESTS")

for case in edge_case_inputs:
    output = preprocess_text(case)
    status = "PASS" if output == [] else "CHECK"
    print(f"\n  Input  : {repr(case)}")
    print(f"  Output : {output}")
    print(f"  Status : {status}")


print("\nAll edge cases handled — no crashes!")
```
