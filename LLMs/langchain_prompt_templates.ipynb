## Task 1: Replace Hardcoded Prompts (10%)

**Goal:** Convert a hardcoded f-string function into a reusable `PromptTemplate`.

**Before (Bad Practice):**
```python
def explain_topic(topic):
    return f"Explain {topic} in simple terms for beginners"  # f-string
```

````python
# ─── Task 1: Replace Hardcoded Prompts ──────────────────────────────────────────

# LangChain PromptTemplate — no f-strings, fully reusable
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# ─── Reusable function using the template ───────────────────────────────────────
def explain_topic(topic: str) -> str:
    """
    Generates a beginner-friendly explanation prompt for any topic.
    Uses PromptTemplate — no f-strings, fully reusable.
    """
    return explain_template.format(topic=topic)

# ─── Test ───────────────────────────────────────────────────────────────────────
print("=" * 55)
print("Task 1: Replace Hardcoded Prompts")
print("=" * 55)

test_topics = ["Photosynthesis", "Blockchain", "Machine Learning"]
for t in test_topics:
    print(f"\n Topic: {t}")
    print("   Prompt:", explain_topic(t))

print("\n Task 1 Complete — PromptTemplate used, no f-strings!")
````

## Task 2: Multi-Input Prompt System (15%)

**Goal:** Build a template that takes multiple inputs: `topic`, `audience`, and `tone`.

**Template:** `Explain {topic} for {audience} in a {tone} tone`

````python
# ─── Task 2: Multi-Input Prompt System ──────────────────────────────────────────

# Multi-input PromptTemplate
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# ─── Reusable function ──────────────────────────────────────────────────────────
def explain_for_audience(topic: str, audience: str, tone: str) -> str:
    """
    Generates a structured prompt for any topic, audience, and tone combination.
    """
    return multi_input_template.format(topic=topic, audience=audience, tone=tone)

# ─── Test Cases ─────────────────────────────────────────────────────────────────
test_cases = [
    {"topic": "AI",            "audience": "beginners",  "tone": "friendly"},
    {"topic": "Python",        "audience": "kids",        "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers",   "tone": "technical"},
]

print("=" * 55)
print("Task 2: Multi-Input Prompt System")
print("=" * 55)

for i, case in enumerate(test_cases, 1):
    prompt = explain_for_audience(**case)
    print(f"\n Test Case {i}:")
    print(f"   Inputs  → topic='{case['topic']}', audience='{case['audience']}', tone='{case['tone']}'")
    print(f"   Prompt  → {prompt}")

print("\n Task 2 Complete — Multi-input template working!")
````

## Task 3: Prompt Variations Engine (15%)

**Goal:** Create 3 distinct `PromptTemplate` styles for the same topic.

| Style | Template |
|-------|----------|
| Teaching | `Explain {topic} clearly step by step` |
| Interview | `Ask 3 questions about {topic}` |
| Storytelling | `Explain {topic} as a story` |

````python
# ─── Task 3: Prompt Variations Engine ───────────────────────────────────────────

# Define 3 separate PromptTemplates for different styles
PROMPT_VARIATIONS = {
    "teaching": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} clearly step by step"
    ),
    "interview": PromptTemplate(
        input_variables=["topic"],
        template="Ask 3 insightful interview questions about {topic}"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} as an engaging and creative story"
    ),
}

# ─── Variations Engine function ─────────────────────────────────────────────────
def run_prompt_variations(topic: str) -> dict:
    """
    Runs all 3 prompt style variations for a given topic.
    Returns a dictionary of style -> generated prompt.
    """
    results = {}
    for style, template in PROMPT_VARIATIONS.items():
        results[style] = template.format(topic=topic)
    return results

# ─── Run for Machine Learning ────────────────────────────────────────────────────
topic = "Machine Learning"
variations = run_prompt_variations(topic)

print("=" * 55)
print(f"Task 3: Prompt Variations Engine — Topic: '{topic}'")
print("=" * 55)

icons = {"teaching": "Teaching", "interview": "Interview", "storytelling": "Storytelling"}
for style, prompt in variations.items():
    print(f"\n Style: {style.upper()}")
    print(f"   {prompt}")

print("\n Task 3 Complete — 3 variation templates working!")
````

## Task 4: ChatPromptTemplate System (15%)

**Goal:** Build a chat-based prompt system with `System` and `User` roles.

| Role | Purpose |
|------|---------|
| System | Defines the AI's behavior/persona |
| User | Dynamic input from the user |

````python
# ─── Task 4: ChatPromptTemplate System ──────────────────────────────────────────

# Role-based system prompt definitions (no hardcoding)
ROLE_SYSTEM_PROMPTS = {
    "teacher": "You are an expert teacher who explains complex topics clearly and patiently.",
    "interviewer": "You are a professional interviewer who asks sharp, insightful questions.",
    "motivator": "You are an enthusiastic motivator who inspires people to learn new things.",
}

# Build ChatPromptTemplate dynamically based on role
def build_chat_prompt(role: str, topic: str) -> list:
    """
    Builds a structured chat prompt using system + human message roles.
    
    Args:
        role  : One of 'teacher', 'interviewer', 'motivator'
        topic : The subject to discuss
    
    Returns:
        List of formatted chat messages (system + human)
    """
    if role not in ROLE_SYSTEM_PROMPTS:
        raise ValueError(f"Invalid role '{role}'. Choose from: {list(ROLE_SYSTEM_PROMPTS.keys())}")

    # Build system message template
    system_template = SystemMessagePromptTemplate.from_template(
        ROLE_SYSTEM_PROMPTS[role]
    )

    # Build human/user message template
    human_template = HumanMessagePromptTemplate.from_template(
        "Please help me understand {topic}."
    )

    # Combine into a ChatPromptTemplate
    chat_prompt = ChatPromptTemplate.from_messages([system_template, human_template])

    # Format and return messages
    return chat_prompt.format_messages(topic=topic)


# ─── Test Cases ─────────────────────────────────────────────────────────────────
test_configs = [
    {"role": "teacher",      "topic": "Neural Networks"},
    {"role": "interviewer",  "topic": "Data Structures"},
    {"role": "motivator",    "topic": "Quantum Computing"},
]

print("=" * 55)
print("Task 4: ChatPromptTemplate System")
print("=" * 55)

for config in test_configs:
    messages = build_chat_prompt(**config)
    print(f"\n Role: {config['role'].upper()} | Topic: {config['topic']}")
    for msg in messages:
        role_label = msg.__class__.__name__.replace("Message", "")
        print(f"   [{role_label}]: {msg.content}")

print("\n Task 4 Complete — ChatPromptTemplate with roles working!")
````

## Task 5: Input Validation Layer (10%)

**Goal:** Create a `validate_inputs()` function that enforces allowed values.

| Field | Allowed Values |
|-------|----------------|
| `audience` | `beginner`, `intermediate`, `expert` |
| `tone` | `formal`, `casual`, `fun` |

````python
# ─── Task 5: Input Validation Layer ─────────────────────────────────────────────

# Allowed values — defined once, reused everywhere
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES     = ["formal", "casual", "fun"]

DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE     = "casual"


def validate_inputs(
    audience: str,
    tone: str,
    use_defaults: bool = True
) -> tuple:
    """
    Validates audience and tone inputs against allowed values.

    Args:
        audience     : Target audience level.
        tone         : Desired tone of explanation.
        use_defaults : If True, assigns defaults on invalid input.
                       If False, raises a ValueError.

    Returns:
        Tuple of (validated_audience, validated_tone)

    Raises:
        ValueError: If inputs are invalid and use_defaults=False.
    """
    audience_clean = audience.strip().lower()
    tone_clean     = tone.strip().lower()

    errors = []

    # ── Validate audience ──
    if audience_clean not in VALID_AUDIENCES:
        if use_defaults:
            print(f"  Invalid audience '{audience}' → defaulting to '{DEFAULT_AUDIENCE}'")
            audience_clean = DEFAULT_AUDIENCE
        else:
            errors.append(
                f"Invalid audience '{audience}'. Allowed: {VALID_AUDIENCES}"
            )

    # ── Validate tone ──
    if tone_clean not in VALID_TONES:
        if use_defaults:
            print(f"  Invalid tone '{tone}' → defaulting to '{DEFAULT_TONE}'")
            tone_clean = DEFAULT_TONE
        else:
            errors.append(
                f"Invalid tone '{tone}'. Allowed: {VALID_TONES}"
            )

    if errors:
        raise ValueError("Validation failed:\n  - " + "\n  - ".join(errors))

    return audience_clean, tone_clean


# ─── Test Cases ─────────────────────────────────────────────────────────────────
print("=" * 55)
print("Task 5: Input Validation Layer")
print("=" * 55)

# Test 1: Valid inputs
print("\n Test 1: Valid inputs")
result = validate_inputs("expert", "formal")
print(f"   Returned: audience='{result[0]}', tone='{result[1]}'")

# Test 2: Invalid inputs — use defaults
print("\n Test 2: Invalid inputs (use_defaults=True)")
result = validate_inputs("newbie", "aggressive", use_defaults=True)
print(f"   Returned: audience='{result[0]}', tone='{result[1]}'")

# Test 3: Invalid inputs — raise error
print("\n Test 3: Invalid inputs (use_defaults=False) — expects ValueError")
try:
    validate_inputs("newbie", "aggressive", use_defaults=False)
except ValueError as e:
    print(f"   Caught ValueError:\n  {e}")

print("\n Task 5 Complete — Validation layer working!")
````

## Task 6: Prompt Generator App (15%)

**Goal:** Create a `generate_prompt()` function combining validation, templates, and style selection.

**Example Output:**  
`Explain Neural Networks for beginners in a fun storytelling style`

````python
# ─── Task 6: Prompt Generator App ───────────────────────────────────────────────

# Style-specific PromptTemplates with full context
STYLE_TEMPLATES = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=(
            "Explain {topic} for {audience} in a {tone} teaching style. "
            "Break it down clearly step by step."
        )
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=(
            "Generate 3 interview questions about {topic} for {audience} "
            "using a {tone} tone."
        )
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=(
            "Explain {topic} for {audience} in a {tone} storytelling style. "
            "Use vivid analogies and a narrative arc."
        )
    ),
}


def generate_prompt(topic: str, audience: str, tone: str, style: str) -> str:
    """
    Full prompt generator pipeline:
      1. Validates audience and tone
      2. Selects the correct PromptTemplate based on style
      3. Formats and returns the final structured prompt

    Args:
        topic    : Subject to explain
        audience : Target audience (beginner / intermediate / expert)
        tone     : Desired tone (formal / casual / fun)
        style    : Prompt style (teaching / interview / storytelling)

    Returns:
        Formatted prompt string
    """
    # Step 1 — Validate inputs
    validated_audience, validated_tone = validate_inputs(audience, tone)

    # Step 2 — Select template
    if style not in STYLE_TEMPLATES:
        raise ValueError(
            f"Invalid style '{style}'. Choose from: {list(STYLE_TEMPLATES.keys())}"
        )
    template = STYLE_TEMPLATES[style]

    # Step 3 — Format and return
    return template.format(
        topic=topic,
        audience=validated_audience,
        tone=validated_tone
    )


# ─── Test Cases ─────────────────────────────────────────────────────────────────
app_tests = [
    {"topic": "Neural Networks",  "audience": "beginner",     "tone": "fun",    "style": "storytelling"},
    {"topic": "Sorting Algorithms","audience": "intermediate", "tone": "casual", "style": "teaching"},
    {"topic": "System Design",    "audience": "expert",       "tone": "formal", "style": "interview"},
    {"topic": "Recursion",        "audience": "NEWBIE",       "tone": "HARSH",  "style": "teaching"},  # tests defaults
]

print("=" * 60)
print("Task 6: Prompt Generator App")
print("=" * 60)

for i, test in enumerate(app_tests, 1):
    print(f"\n Test {i}: topic='{test['topic']}' | audience='{test['audience']}' | tone='{test['tone']}' | style='{test['style']}'")
    prompt = generate_prompt(**test)
    print(f"   Output: {prompt}")

print("\n Task 6 Complete — Full Prompt Generator App working!")
````

## Full Pipeline Demo

**Flow:** `User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output`

````python
# ─── Full Pipeline Demo ──────────────────────────────────────────────────────────

def run_full_pipeline(topic: str, audience: str, tone: str, style: str) -> str:
    """
    Executes the complete Mini Prompt Engine pipeline:

        User Input
            ↓
        Validation Layer  (Task 5)
            ↓
        Prompt Template   (Tasks 1–3)
            ↓
        Dynamic Generation (Task 6)
            ↓
        Structured Output
    """
    print("\n" + "═" * 60)
    print("  Mini Prompt Engine — Full Pipeline")
    print("═" * 60)
    print(f"  INPUT  → topic='{topic}', audience='{audience}', tone='{tone}', style='{style}'")

    # Step 1 — Validation
    print("\n  STEP 1: Input Validation...")
    v_audience, v_tone = validate_inputs(audience, tone)
    print(f"       Validated → audience='{v_audience}', tone='{v_tone}'")

    # Step 2 — Template Selection
    print("\n  STEP 2: Selecting PromptTemplate...")
    if style not in STYLE_TEMPLATES:
        raise ValueError(f"Unknown style: '{style}'")
    print(f"       Style '{style}' template selected")

    # Step 3 — Prompt Generation
    print("\n  STEP 3: Generating Dynamic Prompt...")
    final_prompt = STYLE_TEMPLATES[style].format(
        topic=topic,
        audience=v_audience,
        tone=v_tone
    )

    # Step 4 — Output
    print("\n  OUTPUT:")
    print(f"  ╔══════════════════════════════════════════════════════╗")
    print(f"  ║  {final_prompt[:52]:<52}  ║")
    if len(final_prompt) > 52:
        print(f"  ║  {final_prompt[52:104]:<52}  ║")
    print(f"  ╚══════════════════════════════════════════════════════╝")

    return final_prompt


# Run the full pipeline
result = run_full_pipeline(
    topic    = "Neural Networks",
    audience = "beginner",
    tone     = "fun",
    style    = "storytelling"
)

print("\n Full Pipeline Complete!")
````
