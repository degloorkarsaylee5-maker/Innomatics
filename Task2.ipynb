```python
# Problem 1: Unique Words in a Sentence


sentence=input("Enter sentence: ")
words_list = sentence.split()


#loop to find unique words
unique = []
for words in words_list:
    if words not in unique:
        unique.append(words)

print("Number of unique words:",len(unique))
print(unique)

```

```python
# Problem 2: Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
max_salary = max(employees.values())

# loop for maxi salary
for name, salary in employees.items():
    if salary == max_salary:
        print(name)

```

```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

max_number=numbers[0]
min_number=numbers[0]

#loop for max number 
for i in numbers:
    if i > max_number:
        max_number=i
        print(max_number)

#loop for max number 
for i in numbers:
    if i > min_number:
        min_number=i
        print(min_number)
```

```python
# Problem 4: Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]

# initial count=0
count=0

#loop for adding count
for i in prices:
    if i>1000:
        count+=1
        
print(count)
```

```python
# Problem 5: Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]

# initial count=0
count=0

#loop for count present students
for i in attendance:
    if i=='P':
        count+=1
print("Attendance Percentage:",(count/len(attendance))*100)
```

```python
# Problem 6: Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = []
for i in phone_numbers:
    if i not in unique_numbers:
        unique_numbers.append(i)
print("uniquee phone numbes:",unique_numbers)
```

```python
# Problem 7: Count Character Frequency

sentence2 = input("Enter the sentence: ")

count = {}

for ch in sentence2:
    if ch in count:
        count[ch] += 1
    else:
        count[ch] = 1

print(sentence2)
print(count)
```

```python
# Problem 10: Calculate Average Marks
enter_marks = input("Enter marks separated by space: ")

marks = list(map(int, enter_marks.split()))

sum_of_marks = sum(marks)
avg_of_marks = sum_of_marks / len(marks)

print(marks)
print("Sum:", sum_of_marks)
print("Average:", avg_of_marks)
```
