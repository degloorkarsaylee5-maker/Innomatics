```python
marks = [45, 78, 90, 33, 60]

#creating an empty array to store passed and failed students
passed_students = []
failed_students = []

for i in marks:
    if i >= 50:
        passed_students.append(i)
    else:
        failed_students.append(i)

#by the len function we could find the number of students passed or failed
count_passed = len(passed_students)
count_failed = len(failed_students)

#printing the clear results
print("Number of students passed:", count_passed)
print("Number of students failed:", count_failed)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

#empty array created
cleared_names = []

for name in names:
    name = name.strip()        
    name = name.lower()        
    cleared_names.append(name)

print(cleared_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msgs in messages:
    length = len(msgs)
    print("Message:", msgs)
    print("Length:", length)
    
    if length > 10:
        print("Flag: Message is longer than 10 characters")
    
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for error in logs:
    if error == "ERROR":
        error_count += 1

print("Total error count:", error_count)
```
