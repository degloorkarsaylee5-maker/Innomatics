```python
def analyze_post_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    print("Social Media Engagement Report")
    print("Analyzed By: Saylee Degloorkar")
    print("Total Likes:", total_likes)
    
    # Conditional check for viral status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example real-time data (likes from different hours/platforms)
likes_data = [250, 300, 150, 200, 150]

analyze_post_engagement(likes_data)
```

```python
def check_rainfall(rainfall_data, required_level):
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain
    
    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)
    
    print("Rainfall Analysis Report")
    print("Analyzed By: Saylee Degloorkar")
    print("Average Rainfall:", round(average_rainfall, 2))
    
    # Conditional logic for rainfall status
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example seasonal rainfall data in mm
seasonal_rainfall = [70, 75, 80, 65, 75]
required_rainfall_level = 70

check_rainfall(seasonal_rainfall, required_rainfall_level)
```

```python
def detect_duplicate_accounts(usernames):
    print("Account Duplication Report")
    print("Checked By: Saylee Degloorkar")
    
    unique_usernames = set()
    duplicates_found = False
    
    # Loop through usernames to check duplicates
    for username in usernames:
        if username in unique_usernames:
            duplicates_found = True
            break
        unique_usernames.add(username)
    
    # Conditional output
    if duplicates_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example list of usernames
usernames_list = ["user123", "admin", "guest", "user123", "moderator"]

detect_duplicate_accounts(usernames_list)
```

```python
def filter_premium_crops(crop_prices, premium_threshold):
    print("Crop Price Analysis Report")
    print("Analyzed By: Saylee Degloorkar")
    
    # Filter premium crops using list comprehension
    premium_crops = [price for price in crop_prices if price > premium_threshold]
    
    print("Premium Crops:", premium_crops)


# Example crop prices in ₹
crop_prices_list = [1500, 2500, 1800, 3200, 1200]
premium_threshold = 2000

filter_premium_crops(crop_prices_list, premium_threshold)
```

```python
def check_attendance_eligibility(student_name, attendance_record):
    print("Attendance Eligibility Report")
    print("Checked By: Saylee Degloorkar")
    print("Student Name:", student_name)
    
    # Calculate total classes attended
    total_classes = len(attendance_record)
    attended_classes = 0
    
    for status in attendance_record:
        if status.lower() == "present":
            attended_classes += 1
    
    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    print("Attendance Percentage:", round(attendance_percentage, 2))
    
    # Determine exam eligibility
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example attendance record for a student
student_name = "Rohit Kumar"
attendance_record = ["present", "absent", "present", "present", "present"]

check_attendance_eligibility(student_name, attendance_record)
```

```python
def calculate_electricity_bill(customer_name, units_consumed):
    print("Electricity Bill Report")
    print("Analyzed By: Saylee Degloorkar")
    print("Customer Name:", customer_name)
    print("Units Consumed:", units_consumed)
    
    # Calculate bill based on slabs
    if units_consumed <= 100:
        bill = units_consumed * 3
    elif units_consumed <= 200:
        bill = (100 * 3) + ((units_consumed - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)
    
    # Determine usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    print("Total Bill: ₹", bill)
    print("Usage Classification:", usage_status)
    
    return bill, usage_status


# Example usage
customer_name = "Sunita Joshi"
units_consumed = 250

calculate_electricity_bill(customer_name, units_consumed)
```
